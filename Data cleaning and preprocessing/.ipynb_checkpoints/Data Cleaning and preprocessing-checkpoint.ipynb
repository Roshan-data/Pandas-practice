 Data Cleaning & Preprocessing
 Real-world data is messy. Pandas gives us powerful tools to clean and transform
 data before analysis

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\rosha\Pandas practice\data cleaning.csv")
```

```python
df
```

```python
#Check for Missing Data
df.isnull()          # True for NaNs
```

```python
df.isnull().sum()       # Count missing per column
```

```python
df.dropna()           # with no missing element in it 
```

```python
df.fillna(0)   #Replace NANs with 0
```

```python
 df["Age"].fillna(df["Age"].mean())     # Replace with mean
```

```python
df.ffill()      #forward fill
```

```python
df.bfill()             #backward fill
```

```python
df.duplicated()       #True the duplicate 
```

```python
df.drop_duplicates()         # Remove duplicate rows
```

String Operations with .str
Works like vectorized string methods and returns a pandas Series 
we cant use all string method here 

```python
df["Name"].str.lower()      #Convert all the element into lower case
```

```python
df["City"].str.contains("Delhi", case = "false")          # Checks if 'delhi' is in the city name, case-insensitive
```

```python
type(df["City"].str.contains("Delhi", case = "false")) 
```

```python
df["Email"].str.split("@")       # Outputs a pandas Series where each element is a list 
```
