Data Transformation
Once your data is clean, the next step is to reshape, reformat, and reorder it as
needed for analysis. Pandas gives you plenty of flexible tools to do this

```python
import pandas as pd 
```

```python
df = pd.read_csv(r"C:\Users\rosha\Pandas practice\researchdata.csv")
```

```python
df
```

```python
df.sort_values("Year")        #Sort according to the years
```

```python
df.sort_values(["Year", "IMDb"])   #sort according to year first if year is tie then it will sort according to
                                                IMDb
```

```python
df2 = df.sort_values(["Year", "IMDb"]).copy()    # for safe side use copy() because pandas does not gurantee 
                                                       #return view or copy of it
```

```python
df2
```

```python
df2.reset_index()              # Reset the index and drop the old index
```

```python
df2["Rank"] = df2["IMDb"].rank()        # Default: average method
```

```python
df2
```

```python
df2.rename(columns={"Film": "Movie"}, inplace=True)           # to rename the files
```
